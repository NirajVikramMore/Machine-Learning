# University_towns.txt

```python
university_towns = []
with open('/content/university_towns.txt') as file:
  for line in file:
    if '[edit]' in line:
      state = line
    else:
      university_towns.append((state,line))
```

```python
university_towns[:5]
```

```python
import pandas as pd
```

```python
towns_df = pd.DataFrame(university_towns, columns=['State', 'RegionName'])
towns_df.head()
```

```python
def get_city_state(item):
  if '(' in item:
    return item[:item.find('(')-1]
  elif '[' in item:
    return item[:item.find('[')]
  else:
    return item
```

```python


towns_df = towns_df.applymap(get_city_state)

```

```python
towns_df.head()
```

```python
states = towns_df.groupby('State')['RegionName'].value_counts()
```
