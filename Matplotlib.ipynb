```python
iris = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
iris
```

```python



iris['variety'] = iris['variety'].replace({'Setosa': 0, 'Versicolor': 1, 'Virginica': 1})
iris
```

```python

X = iris[['petal.length']] # Petal length
y = iris[['variety']]# Species: Iris virginica (1) or not (0)

# Train a logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Make predictions on the test set
y_pred = model.predict(X)





```

```python

plt.plot( X,y_pred, color='red')
plt.xlabel('Petal Length')
plt.ylabel('Iris Virginica (1 or 0)')
plt.title('Logistic Regression on Iris Dataset')
plt.show()
```

```python

plt.plot( X,y_pred, color='red')
plt.xlabel('Petal Length')
plt.ylabel('Iris Virginica (1 or 0)')
plt.title('Logistic Regression on Iris Dataset')

plt.grid(True)
plt.figtext(0.995,0.01,'Footnote',ha='right',va='bottom')
plt.legend(loc='best',framealpha=0.5,prop={'size':'small'})
plt.tight_layout(pad=1)
plt.gcf().set_size_inches(8,4)
plt.show()
```

```python
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
size = ((np.random.rand(N)+1)*8)*2;
colours = np.random.rand(N)
fig, ax = plt.subplots(figsize = (8,4))
l = ax.scatter(x,y,s=size,c=colours)
fig.colorbar(l)
ax.set_xlim(-0.05,1.05)
ax.set_ylim(-0.05,1.05)
fig.suptitle('Dramatic scatter plot')
fig.tight_layout(pad=2)
ax.grid(True)
fig.savefig('filename.png', dpi=125)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
markers = list('ov^<>12348sphHdD+x*_')

N = 10

for i, m in enumerate (markers):
  x= np.arange(N)

  y= np.repeat(i+1, N)

  ax.scatter(x, y, marker=m, label=m, s=50, c='cornflowerblue')

ax.set_xlim((-1,N))

ax.set_ylim((0,len(markers)+1))

ax.legend(loc='lower right', ncol=3,prop={'size': 'xx-large'},shadow=True, title='Marker Legend')

ax.get_legend().get_title().set_color("red")

fig.suptitle('Markers ' + '(with an oversized legend)')

fig.tight_layout(pad=2);

fig.savefig('filename.png', dpi=125)
```

```python
N=5
labels = list('ABCDEFGHIJKLM' [0:N])

data = np.array (range(N)) + np.random.rand(N)

#plot the data

fig, ax = plt.subplots(figsize=(8, 3.5))
width=0.8;

tickLocations=  np.arange(N)

rectLocations = tickLocations-(width/2.0)

ax.bar(rectLocations, data, width,color='wheat',edgecolor='yellow', linewidth=4.0)

# pretty-up the plot

ax.set_xticks(ticks = tickLocations)

ax.set_xticklabels (labels)

ax.set_xlim(min(tickLocations)-0.6, max(tickLocations)+0.6)

ax.set_yticks(range (N) [1:])

ax.set_ylim((0,N))

ax.yaxis.grid(True)

# title and save



fig.suptitle("Bar Plot with " + "Oversized Edges")

fig.tight_layout(pad=2)
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
```

```python
df.groupby('variety').count()
```

```python
grp = df.groupby('variety').groups
cat = df['variety'].unique()
count = []

for i in cat :
    count.append({i:len(list(grp.get(i)))})

print(count)
```

```python
grp = df.groupby('variety').groups
cat = df['variety'].unique()
dict = {}

for i in cat :
    dict[i]=len(list(grp.get(i)))
print(dict)
```

```python
labels = list(cat)
plt.bar(labels,list(dict.values()))
plt.title('Count')
```

```python
grp = df.groupby('variety').groups
cat = df['variety'].unique()
dict = {}

for i in cat :
    dict[i]=np.mean(list(grp.get(i)))

labels = list(cat)

plt.bar(labels,list(dict.values()))
plt.title('Mean')
```
